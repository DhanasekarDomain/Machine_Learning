```python
import pandas as pd
```

```python
# pandas is python library process and manipulate tabular data, implementing operations.
# so here we import that pandas (library) as in the name of pd.
```

```python
database=pd.read_csv("Salary_Data.csv")
```

```python
#using the pandas(pd) to read the salary data which is in csv format.
#Then the readed file is assigned in the name of database.
```

```python
database
```

```python
# code for view the data in the "database"
```

```python
independent=database[["YearsExperience"]]
dependent=database[["Salary"]]
```

```python
# we assigned the "YearsExperience" coloumn in "database" to "independent" Variable.
# and assigned the "Salary" coloumn in "database" to "dependent" Variable.
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.10,random_state=0)
```

```python
# "sklearn" is collection of tools that allow you to quickly build and deploy machine learning models in Python
# from sklearn in model_selection importing train_test_split.
# In that using train_test_split function we are assigning value for x_train,x_test,y_train and y_test.
# independent variable for x series, dependent variable for y series. and also 90% of data assigned to "training" to get the more accurate result.
# 10% of data assigned for "testing"
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
```

```python
#Now from sklearn access the linear_model to import the LinearRegression, and LinearRegression assigned as regressor.
```

```python
regressor.fit(x_train,y_train)
```

```python
# In this step fitting the x_train and y_train data in regressor (LinearRegression) for learning.
```

```python
Y_predict=regressor.predict(x_test)
```

```python
# after learning we are giving x_test data to predict the value using regressor. And this is assigned as Y_predict (predicted value)
```

```python
from sklearn.metrics import r2_score
```

```python
# for evaluation we are using sklearn metrics in that importing r2_score.
```

```python
R2_score=r2_score(y_test,Y_predict)
```

```python
# In r2 score, evaluating is done by y_test (actual value) with Y_predict (Model pridict value)
```

```python
R2_score
```

```python
# Code for view the value in R2_score.
# here the value is near to 1 so the model is working perfectly.
```

```python
import pickle
filename="Final_SLR_Model_Assignment.sav"
```

```python
# Pickle is used to converts a Python object (like a list, dictionary, custom class instance, etc.) into a byte stream.
# so import pickle and create a file with defined name.
```

```python
pickle.dump(regressor,open(filename,"wb"))
```

```python
# Here in pickle, we are dumping or sending regressor model. "wb" means write brinary. regressor model written in the form of binary in pickle.
# because pickle is byte stream.
```

```python
load_model=pickle.load(open("Final_SLR_Model_Assignment.sav","rb"))
result=load_model.predict([[15]])
```

```python
# in pickle opening the saved file to read (rb means Read Binary). that file named as load_model.
# to predict the salary for 15 years experience. using the load_model to predic the value and that value assigned name as result.
```
